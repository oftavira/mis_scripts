```python
Ent = []
for i in range(len(Entidades)):
    nuevo_estado = Entidades[i].replace(" ","")
    if nuevo_estado in Ent:
        pass
    else:
        Ent.append(nuevo_estado)
tuple(Ent)
```

```python
Emp = []
for i in range(len(Empresas)):
    nueva_empresa = Empresas[i].replace(" ","").casefold()
    if nueva_empresa in Emp:
        pass
    else:
        Emp.append(nueva_empresa)
Emp.sort()
Emp
```
